```python
list_links = list()
base_link = "https://www.wunderground.com/history/daily/tr/istanbul/LTBA/date/"
for years in range(2020,2023):
    for months in range(1,13):
        for days in range (1,29):
            date_link = base_link+str(years)+"-"+str(months)+"-"+str(days)
            list_links.append(date_link)
```
